## Visualization

```python
#Import matplotlib and seaborn to use for our visualizations
import matplotlib as mp
import seaborn as sns
```

```python
#Setting the seaborn figure size so all the information can fit on the visualizations
sns.set(rc={"figure.dpi":175, 'savefig.dpi':175})
```

What countries pay the most for the basic package?

```python
a = sns.barplot(x='Country_code', y='Cost Per Month - Basic ($)', data=netflix_df, palette = 'Accent')
a.set_xlabel('Country', fontsize=20)
a.set_ylabel('Price', fontsize=20)
a.tick_params(axis='x', rotation=90, labelsize = 7)
```

What about the standard package?

```python
a = sns.barplot(x='Country_code', y='Cost Per Month - Standard ($)', data=netflix_df, palette = 'pastel')
a.set_xlabel('Country', fontsize=20)
a.set_ylabel('Price', fontsize=20)
a.tick_params(axis='x', rotation=90, labelsize = 7)
```

And finally what about the premium package?

```python
a = sns.barplot(x='Country_code', y='Cost Per Month - Premium ($)', data=netflix_df, palette = 'muted')
a.set_xlabel('Country', fontsize=20)
a.set_ylabel('Price', fontsize=20)
a.tick_params(axis='x', rotation=90, labelsize = 7)
```

```python
a = sns.barplot(x='Country_code', y='No. of TV Shows', data=netflix_df, palette = 'muted')
a.set_xlabel('Country', fontsize=20)
a.set_ylabel('Number of Shows', fontsize=20)
a.tick_params(axis='x', rotation=90, labelsize = 7)
```

```python
a = sns.barplot(x='Country_code', y='No. of Movies', data=netflix_df, palette = 'dark')
a.set_xlabel('Country', fontsize=20)
a.set_ylabel('Number of Movies', fontsize=20)
a.tick_params(axis='x', rotation=90, labelsize = 7)
```
